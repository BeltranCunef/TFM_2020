Carga de datos y selección de las variables finalmente usadas en el modelo.

```python
datos = pd.read_csv('../data/datos_procesados.csv')
```

```python
# features
features = list(pd.read_csv('../data/datos_edaspy.csv').columns.values)
features.remove('CODE_GENDER') # no se puede sesgar por genero
features.append('AMT_CREDIT')
features.append('TARGET')
```

```python
# seleccion de features sobre datos
datos = datos[features]
```

```python
# Eliminacion de 12 registros con NAs
datos = datos[datos['AMT_ANNUITY'].notna()]
```

```python
# Establecemos el interes en funcion de la cantidad pedida
def interes(cantidad):
    if((cantidad <= 80000)):
        return 0.06

    elif((cantidad > 80000) & (cantidad <= 150000)):
        return 0.04

    elif((cantidad > 150000) & (cantidad <= 650000)):
        return 0.025

    else:
        return 0.0175
```

```python
# En ciertos registros la anualidad de los pagos es menor a los intereses del préstamo
intereses = -datos['AMT_CREDIT'] * datos['AMT_CREDIT'].map(lambda x: interes(x))

errores = abs(intereses) >= datos['AMT_ANNUITY']

# Elimino los errores de la muestra
datos = datos[abs(intereses) < datos['AMT_ANNUITY']]
```

```python
# Se escogen únicamente las variables que poseen más de un valor distinto
variables= []
for i in datos.columns.values:
    if datos[i].nunique() > 1:
        variables.append(i)

datos = datos[variables]

del variables
```

### DIVISION TRAIN Y TEST

Se divide igualmente el dataset en train y test, la explicabilidad se hará exclusivamente sobre la parte de test. La razón se basa en que para la obtención de la explicabilidad del modelo es necesario la realización de predicciones, por ello, se realizan las divisiones igualmente del dataset utilizando el mismo random state que se usó en su momento para el modelado.

```python
# separamos las variables del target
X = datos.loc[:, datos.columns != 'TARGET']
Y = datos['TARGET']

# dividimos en train y test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=32)
```

### 1. Shap Values

#### 1.1 Comprensión global

```python
import shap
# Need to load JS vis in the notebook
shap.initjs() 
```

```python
%%time
explainer = shap.TreeExplainer(modelo)
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values[1], X_test)
```

##### Interpetación 

En primer lugar, se ha de aclarar que los valores obtenido en los _shap values_ de cada una de las variables explican la correlación de esta con el resultado obtenido como _output_.

A continuación, procederé a la interpretación de los _shap values_ obtenidos.
* Se puede observar que las variables `EXT_SOURCE_2`,`EXT_SOURCE_3` y `EXT_SOURCE_1`  que se obtienen de los datos de _bureau_ , se comportan de forma indirectamente proporcional a la métrica utilizada en la parte de modelado, se observa que cuanto mayor es el valor de estas 3 variables, menor es el _output_ y viceversa.
* Sucede al contrario con otras variables como `AMT_ANNUITY` y `AMT_CREDIT`. En estas, mayores valores representan una mayor repercusión en el _output_.

```python
shap.summary_plot(shap_values[1], X_test, plot_type='bar')
```

##### Interpetación 

En este segundo caso, lo que se observa es la significatividad de cada una de las variables, de mayor a menor importancia, en el _output_ , independientemente de la correlación con el resultado.

* Las variables más relevantes en la obtención del _output_ son `EXT_SOURCE_2`,`EXT_SOURCE_3` y `EXT_SOURCE_1`.

### 2. Shap Values + LGBM

```python
%%time
pd_shap = pd.DataFrame(modelo.predict_proba(X_test, pred_contrib=True), 
                       columns=list(X_test.columns)+['other']).drop(['other'],axis=1)
```

```python
# predicción de TEST
pd_test_predicted = pd.DataFrame(modelo.predict_proba(X_test), 
                                 index=X_test.index, columns = ['y_predicted_0', 'y_predicted']).drop(['y_predicted_0'],axis=1)
pd_test_predicted_final = pd.concat([X_test, pd_test_predicted, Y_test],axis=1)
```

```python
%%time
pd_contrib = pd.DataFrame({'contrib1_max1':['0'], 'contrib1_max2':['0'], 'contrib1_max3':['0'],
                           'contrib0_max1':['0'], 'contrib0_max2':['0'], 'contrib0_max3':['0']})

for i in range(pd_shap.shape[0]):
    list_contrib1 = list(pd_shap.iloc[i,:].sort_values(axis=0, ascending=False).index[-3:])
    list_contrib0 = list(pd_shap.iloc[i,:].sort_values(axis=0, ascending=False).index[:3])
    pd_row = pd.DataFrame([list_contrib1+list_contrib0])
    pd_row.columns=pd_contrib.columns
    pd_contrib = pd.concat([pd_contrib, pd_row],axis=0)
    
pd_contrib = pd_contrib.iloc[1:,].set_index(X_test.index)
pd_contrib = pd.concat([pd_test_predicted_final, pd_contrib],axis=1)
```

```python
pd_contrib.sample(10)
```

En este dataframe podemos observar una serie de 10 registros aleatorios, a la derecha de la tabla se aprecia la probabilidad de _dafault_ predicha por el modelo, el _target_ real y las variables que han contribuido en mayor medida a que esa probabilidad se acerque a 1 (contrib1) y a 0 (contrib0).
